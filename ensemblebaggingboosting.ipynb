```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data")
data
```

```python
x=data.iloc[:,2:]
y=data.iloc[:,1].apply(lambda x:1 if x=='M' else 0)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
adaboost=AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1),n_estimators=200)
adaboost.fit(x_train,y_train)
y_pred=adaboost.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
data=pd.read_csv(url,header=None)
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42)
rf_classifier.fit(x_train,y_train)
```

```python
y_pred=rf_classifier.predict(x_test)
```

```python
accuaracy=accuracy_score(y_test,y_pred)
accuracy
```
