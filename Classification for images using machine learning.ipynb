```python
import torch
import torchvision
import matplotlib.pyplot as plt
from torch import nn
import torch.nn.functional as F
from torchvision import datasets,transforms
from collections import OrderedDict
```

```python
transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,),(0.5,))])
```

```python
mnist_data=datasets.MNIST('mnist_train_data',download=True,train=True,transform=transform)
```

```python
mnist_data_loader=torch.utils.data.DataLoader(mnist_data,batch_size=64,shuffle=True)
```

```python
train_data_iterator=iter(mnist_data_loader)
```

```python
train_images,train_labels=next(train_data_iterator)
```

```python
train_images.shape
```

```python
train_labels.shape
```

```python
train_labels
```

```python
plt.imshow(train_images[63].numpy().squeeze(),cmap='gray')
```

```python
class NeuralNetwork(nn.Module):
  def __init__(self):
    super().__init__()

    self.hidden_layer=nn.Linear(784,256)
    self.output_layer=nn.Linear(256,10)

  def forward(self,x):
    x=F.relu(self.hidden_layer(x))
    x=F.softmax(self.output_layer(x),dim=1)
    return x
```

```python
model=NeuralNetwork()
```

```python
model
```

```python
model.hidden_layer.weight
```

```python
model.hidden_layer.bias
```

```python
model.hidden_layer.weight.data.fill_(0)
```

```python
criterion=nn.CrossEntropyLoss()
```

```python
train_images.shape
```

```python
train_images_batch=train_images.view(train_images.shape[0],-1)
```

```python
train_images_batch.shape
```

```python
output_predictions=model(train_images_batch)
```

```python
loss=criterion(output_predictions,train_labels)
```

```python
loss
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=8
for e in range(epochs):
  training_loss=0
  for train_images,train_labels in mnist_data_loader:
      images=train_images.view(train_images.shape[0],-1)
      optimizer.zero_grad()
      output_predictions=model(images)
      loss=criterion(output_predictions,train_labels)
      loss.backward()
      optimizer.step()
      training_loss+=loss.item()
  else:
    print(f"training loss for epoch - {e}:{training_loss/len(mnist_data_loader)}")
```

```python
from torchvision import models
vgg16=models.vgg16(pretrained=True)
vgg16
```

```python
for param in vgg16.features.parameters():
  param.requires_grad=False
```

```python
import torch.nn as nn
```

```python
vgg16.classifier[6]
```

```python
num_inputs=vgg16.classifier[6].in_features
```

```python
classes = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']
last_fcn_layer=nn.Linear(num_inputs,len(classes))
```

```python
vgg16.classifier[6]=last_fcn_layer
```

```python
vgg16
```

```python
vgg16.cuda()
```

```python
import torch.optim as optim
```

```python
criterion=nn.CrossEntropyLoss()
```

```python
optimizer=optim.SGD(vgg16.classifier.parameters(),lr=0.001)
```

```python
epochs=2
for epoch in range(1,epochs+1):
  traing_loss=0.0
  for b_i,(data,target) in enumerate(train_data_loader):
    data,target=data.cuda(),target.cuda()

    optimizer.zero_grad()

    output=vgg16(data)

    loss=criterion(output,target)

    loss.backward()

    optimizer.step()

    train_loss=loss.item()

    if b_i%20==19:
      print(epoch,b_i+1,train_loss/20)
      train_loss=0.0
```

```python
vgg16.eval()
```

```python
total=0
correct=0
for b_i,(data,target) in enumerate(test_data_loader):
    data,target=data.cuda(),target.cuda()

    optimizer.zero_grad()

    output=vgg16(data)

    _,pred=torch.max(output,1)

    total=target.size(0)

    correct+=(pred==target).sum().item()
```

```python
accuracy=100*correct/total
```
